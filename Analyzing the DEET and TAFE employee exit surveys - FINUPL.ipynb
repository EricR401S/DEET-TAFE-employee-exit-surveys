# <center> Measuring Dissatisfaction Rates from Resigning TAFE and DETE Employees <center>

## <center> Introduction <center>

#### The purpose of this study is to answer one key question from two lens. 
    
> -  Are the employees who have worked in the TAFE and DETE departments for a longer or shorter timeframe resigning due to some kind of dissatisfaction?
    
> -  Are the younger ones resigning due some dissatisfaction? What about the older ones? 
    
#### This study is based on the [dataset](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey) from the Department of Education, Training and Employment (DETE) and [another](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) from the Technical and Further Education (TAFE), in which both are exit survey data. The key questions, in their bare essentials, are reduced to one. The employers wants to know if the employees who resign are leaving dissatisfied, first and foremost, and, if so, how does that dissatisfaction rate fluctuate by age or career stage.This is the first purpose of this study, while the second is for course training in data cleaning.   

## <center> Summary of Results <center>
    
#### The datasets were cleaned and modified to better identify the age and tenure. After classifying the employees who quit as dissatisfied or not, the following results were calculated and obtained.
    
    
| Tenure      | Dissatisfaction % |
|-------------|-------------------|
| New         | 47%               |
| Experienced | 58%               |
| Established | 77%               |
| Veteran     | 80%               |
| All         | 62%               |    

    
| Age            | Dissatisfaction % |
|----------------|-------------------|
| 30 and younger | 61%               |
| 31 to 40       | 64%               |
| 41 to 50       | 65%               |
| 51 and older   | 65%               |
| All            | 64%               |
   
####  
    

#### As the results demonstrate, slightly more than 60% of the employees who resign are dissatisfied upon exiting, both from the lens of tenure (62%) and age (64%). From the career stage metrics, dissatisfaction was rather high for the established (7 to 10 years) and veteran (more than 11 years) groups, which indicates a potential issues with the options for growth that employees have. As for the age related metrics, 45% of dissatisfied, young employees who quit already had another job for them, a sign of their high mobility and possible lack of growth within their department. In addition, it is also alarming to see a dissatisfaction rate of 65% of employees in their 50s and older who quit their jobs, which leads to questions about company culture. Lastly, it is reccomended for the TAFE and DETE departments to examine access for employee opportunities and the execution of the company culture, so that they may have higher satisfaction rates.

### Identifying Dissatisfaction in the Datasets

####  In the contributing factor columns, the TAFE dataset has confirming values as text, hyphens as a negation, and null values. Therefore, they have to be mapped to "True", "False", and "NaN" to better match the same pattern in the DETE dataset. Then, an overall "dissatisfied" column can be initialized in each dataset if at least one of those contributing factor columns confirms a dissatisfaction, which can be done via the "df.any()" command. In short, the new "dissatisfied" column would confirm the employee left due to some kind of dissatisfaction with the workplace. 

```python
import numpy as np

#Function for changing values in TAFE columns to better match 
#the DETE's contributing factor columns' boolean format

def update_vals(element):
    
    if element == '-':
        
        return False
    
    elif pd.isnull(element):
        
        return np.nan
    
    else:
        
        return True
    
tafe_dissatisfaction_cols = (
    ['dissatisfaction', 'job_dissatisfaction'])

dissatisfaction_values = (
    tafe_resignations[tafe_dissatisfaction_cols]
    .applymap(update_vals))

tafe_resignations['dissatisfied'] = (
    dissatisfaction_values.any(axis = 1,skipna=False))

print(tafe_resignations['dissatisfied'].value_counts(dropna=False))

tafe_resignations_up = tafe_resignations.copy()
```

```python
#DETE
dete_dissatisfaction_cols = ['job_dissatisfaction',
                             'dissatisfaction_with_the_department',
                             'physical_work_environment', 
                             'lack_of_recognition',
                             'lack_of_job_security',
                             'work_location', 
                             'employment_conditions',
                             'work_life_balance', 
                             'workload']

dete_resignations['dissatisfied'] = (
dete_resignations[dete_dissatisfaction_cols]
    .applymap(update_vals).any(axis = 1,skipna=False))


print(dete_resignations['dissatisfied'].value_counts(dropna=False))

dete_resignations_up = dete_resignations.copy()
```

### Dissatisfaction Insights from the Lens of Tenure and Age

#### With the datasets together, the focus can now be shifted to the "institute service" column, which contains the tenures of the employees. The column already carries challenging characteristics due to tenure ranges instead of exact years, such as "7-10". Nevertheless, the approach will be to convert all numbers into categories.

```python
#Gauging the types of values

combined['institute_service'].value_counts(dropna=False)
```

```python
#Function to transform the years into tenure categories

combined['institute_service'] = (
    combined['institute_service'].astype('str'))

#For years in ranges, such as 7-10, only the first year is needed.

pat = r"(?P<years_of_service>[\d]{1,2})" #None will exceed 2 digits

#The following holds the years to be mapped.
years_of_service = (combined['institute_service']
                    .str.extract(pat).astype("float")) 


def tenure_cat(x):
    
    if pd.isnull(x):
        
        return np.nan
    
    elif x < 3:
        
        return "New"
    
    
    elif x >= 3 and x <= 6:
        
        return "Experienced"
    
    
    elif x >= 7 and x <= 10:
        
        return "Established"
    
    
    elif x >= 11:
        
        return "Veteran"
    
combined['service_cat'] = (
    years_of_service['years_of_service'].apply(tenure_cat))

combined['service_cat'].value_counts(dropna=False)
```

The service tenure is one of the two major lens, through which dissatisfaction is being viewed, for this study. To clean this column in the combined dataset, the values had to be stratified according to the [stages](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage) (the article is linked) in which employees find themselves. For this reason, the categories were organized in the previous manner, which additionally addressed the ambiguous, tenure values such as "7-10" and "11-20". Finally, with the tenure data cleaned, an analysis can actually be done on the dissatisfaction from the lens of tenure or service years.

```python
combined["dissatisfied"].value_counts(dropna=False)
```

```python
#Filling the NaNs so calculations can be run

combined["dissatisfied"] = combined["dissatisfied"].fillna(True)

combined["dissatisfied"].value_counts(dropna=False)
```

```python
pv_tenure = (
    combined.pivot_table('dissatisfied','service_cat',
                         aggfunc='mean', margins=True))

pv_tenure_count = (
    combined.pivot_table('dissatisfied','service_cat',
                         aggfunc='count', margins=True)
                         ['dissatisfied'][-1])

import matplotlib.pyplot as plt
%matplotlib inline

print('Count of records with tenure data : ', pv_tenure_count)

percentage_variant = pv_tenure['dissatisfied']*100

percentage_variant.plot(kind='bar', 
                        title = 'Dissatisfied Tenure Group Percentages', 
                        ylim = (0,100))

print('\n',percentage_variant)
```

#### All of this data cleaning has begun to yield fruits. Before elaborating on the insights, it is important to remember that the percentages observed are based on 563 records with tenure data (651 minus 88 NaNs). 

#### The most meaningful insight is that 62% of the 563 employees with tenure data reported feeling dissatisfied. Firstly, "New" and "Experienced" groups numbered a little under and a little over 50% respectively. On the other hand, at a substantially higher rate, 77% of the "Established" and 80% of the "Veteran"group were dissatisfied. 

#### These details are very important. Half of the newer employees already felt inclined to leave, while 80% of the more established ones felt unadressed dissatisfactions, due to such an overwhelming number. In essence, this is an introductory but important analysis under the lens of tenure. What follows is looking at dissatisfaction rates under the scope of age.

```python
combined["age"].value_counts(dropna=False)
```

```python
combined['age'] = combined['age'].astype('str')

#For years in ranges, such as 21-25, only the first year is needed.

pat = r"(?P<age>[\d]{1,2})" #Same as before

employee_ages = (combined['age']
                    .str.extract(pat).astype("float"))



def age_cat(x):
    
    if pd.isnull(x):
        
        return np.nan
    
    
    elif x <= 30: #26 to 30
        
        return "30 and younger"
    
    
    elif x >=31  and x <= 40:
        
        return "31 to 40"
    
    
    elif x >=41  and x <= 50:
        
        return "41 to 50"

    
    elif x >= 51:
        
        return "51 and older"

#The choice in category names was to reflect an idea of 
#employees in their 20s, 30s, 40s, 50s and so on.
    
combined['age_cat'] = employee_ages['age'].apply(age_cat)

combined['age_cat'].value_counts(dropna=False)
```

```python
pv_age = (
    combined.pivot_table('dissatisfied','age_cat',
                         aggfunc='mean', margins=True))

pv_age_count = (
    combined.pivot_table('dissatisfied','age_cat',
                         aggfunc='count', margins=True)
                         ['dissatisfied'][-1])

print('Count of records with age data : ', pv_age_count)

age_percentage_variant = pv_age['dissatisfied']*100

age_percentage_variant.plot(kind='bar', 
                            title = 'Dissatisfied Age Group Percentages', 
                            ylim = (0,100))

print('\n',age_percentage_variant)
```

#### Once again, before expounding on dissatisfaction by age groups, it is important to reitirate that the percentages here are based on 596 records with age data (651 minus 55 NaNs). 

#### Across every group, 60% of its members reported leaving dissatisfied. This data is alarming for one reason, the life stages. First, the age groups were divided into their 20s, 30s, 40s, and 50s or more. In other words, they were divided according to life stages, not career stages. The 20s are years of high mobility, and, in similar fashion, the 30s share some traits, but are more promotion minded. This mindset carries into the 40s and 50s (and older), and, furthermore, these two groups are thinking about retirement. These details tend to be common knowledge, and marriage and children also tend to influence staying at a given workplace for stability.

#### This line of thinking leads to the following inferences, 60% of employees in their 20s, 30s, and 40s felt  dissatisfaction in their workplaces, probably indicative of low growth. More interestingly, 60% of employees in their 50s and older, felt dissatsified enough to forego their future benefits and resign. The latter group, whose members typically settle around this age, not just left, but left unhappy. ### Career Stage Inferences : A Further Look

#### Lack of growth or opportunity isn't exactly a column of data that was well defined or shared across the DETE and TAFE datasets. However, one of the major clues is if the employees moved to another job. First, it's important to gauge how many of the employees are dissatisfied in total.

```python
combined['dissatisfied'].value_counts(dropna=False)
```

#### With 410 dissatisfied employees, the next step is to produce insights from this subgroup from the two main perspectives, tenure and age. The following steps will be to clean, prepare and plot the data.

```python
#Implementing a variant of a previous function

#The values of these two columns before the mapping
print(combined['career_move_to_private_sector']
      .value_counts(dropna=False),'\n')

print(combined['career_move_to_public_sector']
      .value_counts(dropna=False),'\n')


career_cols =['career_move_to_public_sector', 
               'career_move_to_private_sector']
    

def update_career_cols(val):
    
    if val == '-' or val == False:
        
        #The columns have some boolean values, see results.
        
        return False 
    
    elif pd.isnull(val):
        
        return np.nan
    
    else:
        
        return True
    
for col in career_cols:
    
    combined[col] = combined[col].map(update_career_cols)
    
#Value counts after the mapping
print(combined['career_move_to_private_sector']
      .value_counts(dropna=False),'\n')

(combined['career_move_to_public_sector']
 .value_counts(dropna=False),'\n')
```

```python
#Creating a dataframe for calculations based on job change

dissatisfaction_boolean = combined['dissatisfied']==True

dissatisfied_employees = combined[dissatisfaction_boolean]


#the main categories and their denominators, used for plotting later

serv_cat_denominator = (
    dissatisfied_employees
    .pivot_table('dissatisfied','service_cat',aggfunc = np.sum, 
                 margins = True))

serv_cat_denominator = serv_cat_denominator['dissatisfied']

age_cat_denominator = (
    dissatisfied_employees
    .pivot_table('dissatisfied','age_cat', aggfunc = np.sum, 
                 margins = True))

age_cat_denominator = age_cat_denominator['dissatisfied']


#The resulting denominators

print(serv_cat_denominator, '\n')

age_cat_denominator
```

```python
serv_job_change = (dissatisfied_employees
     .pivot_table(career_cols,'service_cat', 
                  aggfunc = np.sum, margins = True))

age_job_change = (dissatisfied_employees
     .pivot_table(career_cols,'age_cat', 
                  aggfunc = np.sum, margins = True))


serv_job_change['Total'] = (
    serv_job_change['career_move_to_private_sector'] 
    + serv_job_change['career_move_to_public_sector'])

age_job_change['Total'] = (
    age_job_change['career_move_to_private_sector']
    + age_job_change['career_move_to_public_sector'])

#Appending a 'Total' to this list for the following 'for' loop
career_cols.append('Total')

for col in career_cols: #To get percentages
    
    serv_job_change[col] = ((serv_job_change[col] 
                             / 
                             serv_cat_denominator) * 100)
    
    age_job_change[col] = ((age_job_change[col] 
                            /
                            age_cat_denominator) * 100)
    
```

```python
print('Service Category Denominator : ', 
      serv_cat_denominator[-1])


serv_job_change.plot(kind='bar',
       title = 'Dissatisfied Employees Who Quit and Changed Jobs by Service Category',
       ylim = (0,80))

serv_job_change
```
